#### Discussions:
- Discuss types of data (discrete, continuous).
- Discuss why we differentiate between population and sample in statistics.

## Answer:
### * Discrete data and continuous data:
#### Discrete data and continuous data are both quantitative data type.
#### Discrete data: The type of quantitative data that has discrete integer values that we can count. For example, Number of cars in a parking lot, Number of students present in the class, etc.

#### Continuous data: The type of quantitative data that has values as real numbers. For example, Temperature data of a day from morning to midnight, Height of students in a class can be 165.1 cm, 170.45 cm, 188.6 cm, etc, are continuous data

#### * Population and Sample in statistics

#### Population is the entire data set collected for study of a particular feature. For example, collecting data of heights of all people living in India will be referred to as population data set of height in India. 

#### Sample is a subset of Population. Sample is the data collected which can be used to deduce the features in a population. For example, collecting data of heights of 1000 people from all states of India is a sample data. We can calculate the statistical parameters like mean, median, etc from this sample to deduce the information of population.

### Descriptive Statistical Measures
Explore key measures like central tendency (mean, median, mode) and dispersion (range, variance, standard deviation).

- Measures of central tendency: mean, median, mode.
- Measures of dispersion: range, variance, standard deviation, and interquartile range (IQR).

- Calculate range, variance, and standard deviation.
- Calculate IQR for understanding data spread.

```python
data_df['gender'].unique()
```

```python
# Mean, Median, Mode 

fig, axes = plt.subplots(1,2, figsize=(15,6))

m=data_df[['gender','math score']].groupby(['gender'],as_index=False).mean()
m['gender'] = m['gender'].replace({0: 'Female', 1: 'Male'})
m.plot(x='gender',y='math score',kind='bar',ax=axes[0], color='k')


m=data_df[['gender','math score']].groupby(['gender'],as_index=False).median()
m['gender'] = m['gender'].replace({0: 'Female', 1: 'Male'})
m.plot(x='gender',y='math score',kind='bar',ax=axes[1], color='k')
```

```python
data_df.describe()
```

```python
# Range, Variance, Standard Deviation

for i in num_col:
    print(f"Range of '{i}': {data_df[i].max()-data_df[i].min()}")

print()
for i in num_col:
    print(f"Variance of '{i}': {data_df[i].var().round(2)}")

print()
for i in num_col:
    print(f"Standard Deviation of '{i}': {data_df[i].std().round(2)}")
```

```python
# IQR (Interquartile Range)
for i in num_col:
    Q3 = np.quantile(data_df[i], 0.75)
    Q1 = np.quantile(data_df[i], 0.25)
    IQR = Q3 - Q1

    print(f"Inter Quartile Range of '{i}': {IQR}")
```

### Central Tendency & Dispersion

Dive deeper into calculating and interpreting the central tendency (mean, median, mode) and measures of dispersion (range, variance, standard deviation, quartiles, percentiles).

#### Key Concepts:
- Central tendency: mean, median, mode.
- Dispersion: range, variance, standard deviation, interquartile range.
- Percentiles and Quartiles.
####  Exercise:
- Calculate quartiles and percentiles.
- Visualize quartiles using box plots.
- Compute the coefficient of variation to measure relative variability.

```python
# Calculating Quartiles and Percentiles

for i in num_col:
    q1 = np.percentile(data_df[i],25) # 1st quartile
    q3 = np.percentile(data_df[i],75) # 2nd quartile
    
    print(f"1st Quartile of '{i}' = {q1}")
    print(f"2nd Quartile of '{i}' = {q3}")
    
    iqr = q3-q1
    print(f"Inter Quartile Range of '{i}' = {iqr}")
    print()
```

```python

# Visualizing data using boxplot
# IQR (Interquartile Range)
fig, axes = plt.subplots(3,1, figsize=(10,8))

for count, col in enumerate(num_col):
    ax = axes[count]
    sns.boxplot(x=col, data=data_df, ax = ax, color='c')

plt.tight_layout()
plt.show()
```

```python
num_col
```

```python

# Coefficient of Variation (CV)

cv_dict = {}

for i in num_col:
    a=(data_df[i].std()/data_df[i].mean())*100
    print(f"Coefficient of Variation of '{i}': {a.round(2)}")
    cv_dict[i]=a

sns.barplot(cv_dict,color='darkblue')
plt.ylabel('Coefficient of Variation')
plt.show()
```
